```python
! pip install requests
import requests
req = requests.get('https://timesofindia.indiatimes.com/india/nupur-sharma-case-personal-attacks-on-judges-for-their-judgements-lead-to-a-dangerous-scenario-sc-judge-justice-jb-pardiwala-says/articleshow/92635912.cms')
req
req.status_code
req.content
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(req.content, 'html.parser')
print(soup.prettify())
```

```python
news_title = soup.find_all('h1', class_='_1Y-96')
news_title[0]
title = news_title[0].get_text()
print(title)
```

```python
news_text =soup.find('div',class_='fewcent-92635912')
text = news_text.get_text()
print(text)
```

```python
import csv
data_list = [["Title", "Text"],
            [title, text]]
with open('news1.csv', 'w') as file:
   writer = csv.writer(file)
   writer.writerows(data_list)
```
